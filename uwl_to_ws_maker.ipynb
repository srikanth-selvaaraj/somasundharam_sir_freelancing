```python
!pip install pandas
import pandas as pd
import os
from google.colab import files

arr = os.listdir('/content/')

print('Upload the unique word list file')
uploaded = files.upload()
master_name = next(iter(uploaded))
excel_file = master_name
uwl_df = pd.read_excel(excel_file)
uwl_df = uwl_df.applymap(lambda s:s.lower() if type(s)==str else s)
uwl_df = uwl_df.applymap(lambda s:s.strip() if type(s)==str else s)

if 'master.xlsx' in arr:
  excel_file2 = 'master.xlsx'
  master_df = pd.read_excel(excel_file2)
  master_df_copy = master_df.copy() # for maintaining the cap letters
  master_df = master_df.applymap(lambda s:s.lower() if type(s)==str else s)
  master_df = master_df.applymap(lambda s:s.strip() if type(s)==str else s)
else:
  print('Upload the master list file')
  uploaded2 = files.upload()
  master_name2 = next(iter(uploaded2))
  excel_file2 = master_name2
  master_df = pd.read_excel(excel_file2)
  master_df_copy = master_df.copy() # for maintaining the cap letters
  master_df = master_df.applymap(lambda s:s.lower() if type(s)==str else s)
  master_df = master_df.applymap(lambda s:s.strip() if type(s)==str else s)


col_names = list(uwl_df.columns)
column_index = 0
while column_index < len(col_names):
  col_name = col_names[column_index]
  print(" ")
  print('Processing...' + col_name)

  #Synonyms
  founded_synonyms = pd.DataFrame()
  unfounded_synonyms = []
  df_synonyms = pd.DataFrame() #if there is no founded name it will deliver an empty dataframe otherwise it deliver an error "is not defined"

  for a in uwl_df[col_name]:
      b = master_df[(master_df['SYNONYM'].isin([a])) | (master_df['Unnamed: 1'].isin([a]))]
      if a in b.values:
          founded_synonyms = pd.concat([b,founded_synonyms],ignore_index=True, sort=False)
          founded_synonyms = founded_synonyms.loc[:,['SYNONYM','Unnamed: 1']] # selecting only wanted columns
          founded_synonyms.drop_duplicates(subset="SYNONYM",keep='first', inplace=True)
      else:
          unfounded_synonyms.append(a)

  list_1 = list(founded_synonyms['SYNONYM'])
  list_2 = list(founded_synonyms['Unnamed: 1'])

  for a in list_1:
    if a in list_2:
      if list_1[list_2.index(a)] == list_2[list_1.index(a)]:
        list_1.remove(list_1[list_2.index(a)])
        list_2.remove(list_2[list_2.index(a)])
    else:
      pass

  column_1 = pd.DataFrame(list_1)
  column_2 = pd.DataFrame(list_2)
  founded_synonyms = pd.concat([column_1,column_2],axis=1)

  if founded_synonyms.empty:
      final_df = pd.DataFrame(['There is no synonyms'])
      final_df.columns = ['Result']
      unfounded_df = pd.DataFrame(unfounded_synonyms)
      unfounded_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      final_synonyms_ans = pd.concat([final_df,unfounded_df],axis=1)
  elif len(unfounded_synonyms) == 0:
      final_df = pd.DataFrame(['There is no unfounded words'])
      final_df.columns = ['Result']
      final_synonyms_ans = pd.concat([founded_synonyms,final_df],axis=1)
  else:
      unfounded_df = pd.DataFrame(unfounded_synonyms)
      unfounded_df.columns = ['Unfounded'] # giving column name to unfounded list
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      founded_synonyms.drop_duplicates(keep='first', inplace=True)
      final_df = pd.concat([founded_synonyms,unfounded_df], axis=1)
      final_df.columns = ['SYNONYMS MASTERLIST','SYNONYMS MASTERLIST: Ans','Unfounded']
      final_synonyms_ans = pd.DataFrame(final_df)

  print("synonyms finished")

  #Antonyms picking
  founded_antonyms = pd.DataFrame()
  unfounded_antonyms = []
  df_antonyms = pd.DataFrame()

  for c in uwl_df[col_name]:
      d = master_df[(master_df['ANTONYM'].isin([c])) | (master_df['Unnamed: 3'].isin([c]))]
      if c in d.values:
          founded_antonyms = pd.concat([d,founded_antonyms],ignore_index=True, sort=False)
          founded_antonyms = founded_antonyms.loc[:,['ANTONYM','Unnamed: 3']]
          founded_antonyms.drop_duplicates(subset="ANTONYM",keep='first', inplace=True)
      else:
          unfounded_antonyms.append(c)

  list_1 = list(founded_antonyms['ANTONYM'])
  list_2 = list(founded_antonyms['Unnamed: 3'])

  for a in list_1:
    if a in list_2:
      if list_1[list_2.index(a)] == list_2[list_1.index(a)]:
        list_1.remove(list_1[list_2.index(a)])
        list_2.remove(list_2[list_2.index(a)])
    else:
      pass

  column_1 = pd.DataFrame(list_1)
  column_2 = pd.DataFrame(list_2)
  founded_antonyms = pd.concat([column_1,column_2],axis=1)

  if founded_antonyms.empty:
      unfounded_antonyms_df = pd.DataFrame(unfounded_antonyms)
      unfounded_antonyms_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      final_antonyms_df = pd.DataFrame(['There is no antonyms'])
      final_antonyms_df.columns = ['Result']
      final_antonyms_ans = pd.concat([final_antonyms_df,unfounded_antonyms_df],axis=1)
  elif len(unfounded_antonyms) == 0:
      final_antonyms_df = pd.DataFrame(['There is no unfounded words'])
      final_antonyms_df.columns = ['Result']
      final_antonyms_ans = pd.concat([founded_antonyms,final_antonyms_df],axis=1)
  else:
      unfounded_antonyms_df = pd.DataFrame(unfounded_antonyms)
      unfounded_antonyms_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      founded_antonyms.drop_duplicates(keep='first', inplace=True)
      final_antonyms_df = pd.concat([founded_antonyms,unfounded_antonyms_df], axis=1)
      final_antonyms_df.columns = ['ANTONYM MASTER LIST','ANTONYM MASTER LIST: Ans','Unfounded']
      final_antonyms_ans = pd.DataFrame(final_antonyms_df)

  print("antonym finished")

  #homophones
  founded_homophone = pd.DataFrame()
  unfounded_homophone = []
  df_homophone = pd.DataFrame()

  for g in uwl_df[col_name]:
      h = master_df[(master_df['HOMOPHONE'].isin([g])) | (master_df['Unnamed: 5'].isin([g]))] #searchin a word in any of these columns usin OR login (symbol "|" )
      if g in h.values:
          founded_homophone = pd.concat([h,founded_homophone],ignore_index=True, sort=False)
          founded_homophone = founded_homophone.loc[:,['HOMOPHONE','Unnamed: 5']]
          founded_homophone.drop_duplicates(subset="HOMOPHONE",keep='first', inplace=True)
      else:
          unfounded_homophone.append(g)
  list_1 = list(founded_homophone['HOMOPHONE'])
  list_2 = list(founded_homophone['Unnamed: 5'])

  for a in list_1:
    if a in list_2:
      if list_1[list_2.index(a)] == list_2[list_1.index(a)]:
        list_1.remove(list_1[list_2.index(a)])
        list_2.remove(list_2[list_2.index(a)])
    else:
      pass

  column_1 = pd.DataFrame(list_1)
  column_2 = pd.DataFrame(list_2)
  founded_homophone = pd.concat([column_1,column_2],axis=1)

  if founded_homophone.empty:
      unfounded_homophone_df = pd.DataFrame(unfounded_homophone)
      unfounded_homophone_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      final_homophone_df = pd.DataFrame(['There is no homophones'])
      final_homophone_df.columns = ['Result']
      final_homophone_ans = pd.concat([final_homophone_df,unfounded_homophone_df],axis=1)
  elif len(unfounded_homophone)==0:
      final_homophone_df = pd.DataFrame(['Ther is no unfounded words'])
      final_homophone_df.collumns = ['Result']
      final_homophone_ans = pd.concat([founded_homophone,final_homophone_df],axis=1)
  else:
      unfounded_homophone_df = pd.DataFrame(unfounded_homophone)
      unfounded_homophone_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      founded_homophone.drop_duplicates(keep='first', inplace=True)
      final_homophone_df = pd.concat([founded_homophone,unfounded_homophone_df], axis=1)
      final_homophone_df.columns = ['HOMOPHONES LIST','HOMOPHONES LIST: Ans','Unfounded']
      final_homophone_ans = pd.DataFrame(final_homophone_df)

  print('Homophones finished')


  # compound words
  founded_compound = pd.DataFrame()
  unfounded_compound = []
  master_compound_df = master_df[['COMPOUND', 'Unnamed: 7', 'Unnamed: 8']]

  df_compound = master_compound_df[(master_compound_df['Unnamed: 7'].isin(uwl_df[col_name])) & (master_compound_df['Unnamed: 8'].isin(uwl_df[col_name]))]
  unfounded_compound = uwl_df[(~uwl_df[col_name].isin(df_compound['Unnamed: 7'])) & (~uwl_df[col_name].isin(df_compound['Unnamed: 8']))][col_name]

  if df_compound.empty:
      unfounded_compound_df = pd.DataFrame(unfounded_compound)
      unfounded_compound_df.columns = ['Unfounded']
      final_compound_df = pd.DataFrame(['There is no compound words'])
      final_compound_df.columns = ['Result']
      final_compound_ans = pd.concat([final_compound_df,unfounded_compound_df],axis=1)
  elif len(unfounded_compound) == 0:
      final_compound_df = pd.DataFrame(['There is no unfounded words'])
      final_compound_df.columns = ['Result']
      final_compound_ans = pd.concat([df_compound,final_compound_df],axis=1)
  else:
      unfounded_compound_df = pd.DataFrame(unfounded_compound)
      unfounded_compound_df.columns = ['Unfounded']
      df_compound.reset_index(drop=True, inplace=True)
      unfounded_compound_df.reset_index(drop=True, inplace=True)
      final_compound_df = pd.concat([df_compound, unfounded_compound_df], axis=1)
      final_compound_ans =  pd.DataFrame(final_compound_df)

  print('compound words finished')

  #Meaning
  founded_meaning = pd.DataFrame()
  unfounded_meaning = []
  df_meaning = pd.DataFrame()

  for g in uwl_df[col_name]:
      h = master_df[(master_df['MEANING'].isin([g])) | (master_df['Unnamed: 10'].isin([g]))] #searchin a word in any of these columns usin OR login (symbol "|" )
      if g in h.values:
          founded_meaning = pd.concat([h,founded_meaning],ignore_index=True, sort=False)
          founded_meaning = founded_meaning.loc[:,['MEANING','Unnamed: 10']]
          founded_meaning.drop_duplicates(subset="MEANING",keep='first', inplace=True)
      else:
          unfounded_meaning.append(g)

  if not founded_meaning.empty:
   list_1 = list(founded_meaning['MEANING'])
   list_2 = list(founded_meaning['Unnamed: 10'])
   for a in list_1:
     if a in list_2:
       if list_1[list_2.index(a)] == list_2[list_1.index(a)]:
         list_1.remove(list_1[list_2.index(a)])
         list_2.remove(list_2[list_2.index(a)])
         column_1 = pd.DataFrame(list_1)
         column_2 = pd.DataFrame(list_2)
         founded_meaning = pd.concat([column_1,column_2],axis=1)
     else:
       pass


  if founded_meaning.empty:
      unfounded_meaning_df = pd.DataFrame(unfounded_meaning)
      unfounded_meaning_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      final_meaning_df = pd.DataFrame(['There is no meanings'])
      final_meaning_df.columns = ['Result']
      final_meaning_ans = pd.concat([final_meaning_df,unfounded_meaning_df],axis=1)
  elif len(unfounded_meaning)==0:
      final_meaning_df = pd.DataFrame(['Ther is no unfounded meaning'])
      final_meaning_df.columns = ['Result']
      final_meaning_ans = pd.concat([founded_meaning,final_meaning_df],axis=1)
  else:
      unfounded_meaning_df = pd.DataFrame(unfounded_meaning)
      unfounded_meaning_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      founded_meaning.drop_duplicates(keep='first', inplace=True)
      final_meaning_df = pd.concat([founded_meaning,unfounded_meaning_df], axis=1)
      final_meaning_df.columns = ['MEANING','MEANING: Ans','Unfounded']
      final_meaning_ans = pd.DataFrame(final_meaning_df)

  print('Meanings finished')

  #word - sentence
  founded_sentence = pd.DataFrame()
  unfounded_sentence = []
  df_sentence = pd.DataFrame()

  for g in uwl_df[col_name]:
      h = master_df_copy[master_df['WORD'].isin([g])] # searchin a word in any of these columns usin OR logic (symbol "|" )
      if g in h.values:
          founded_sentence = pd.concat([h,founded_sentence],ignore_index=True, sort=False)
          founded_sentence = founded_sentence.loc[:,['WORD','SENTENCE']]
          founded_sentence.drop_duplicates(subset="WORD",keep=False, inplace=True)
      else:
          unfounded_sentence.append(g)

  if not founded_sentence.empty:
    list_1 = list(founded_sentence['WORD'])
    list_2 = list(founded_sentence['SENTENCE'])
    for a in list_1:
      if a in list_2:
        if list_1[list_2.index(a)] == list_2[list_1.index(a)]:
          list_1.remove(list_1[list_2.index(a)])
          list_2.remove(list_2[list_2.index(a)])
      else:
        pass
    column_1 = pd.DataFrame(list_1)
    column_2 = pd.DataFrame(list_2)
    founded_sentence = pd.concat([column_1,column_2],axis=1)

  if founded_sentence.empty:
      unfounded_sentence_df = pd.DataFrame(unfounded_sentence)
      unfounded_sentence_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded",keep='first', inplace=True)
      final_sentence_df = pd.DataFrame(['There is no words'])
      final_sentence_df.columns = ['Result']
      final_sentence_ans = pd.concat([final_sentence_df,unfounded_sentence_df],axis=1)
  elif len(unfounded_sentence)==0:
      final_sentence_df = pd.DataFrame(['Ther is no unfounded sentence'])
      final_sentence_df.collumns = ['Result']
      final_sentence_ans = pd.concat([founded_sentence,final_sentence_df],axis=1)
  else:
      unfounded_sentence_df = pd.DataFrame(unfounded_sentence)
      unfounded_sentence_df.columns = ['Unfounded']
      unfounded_df.drop_duplicates(subset="Unfounded", keep='first', inplace=True)
      founded_sentence.drop_duplicates(keep='first', inplace=True)
      final_sentence_df = pd.concat([founded_sentence,unfounded_sentence_df], axis=1)
      final_sentence_df.columns = ['WORD','SENTENCE','Unfounded']
      final_sentence_ans = pd.DataFrame(final_sentence_df)

  print('Word sentence finished')


  all_df = [final_synonyms_ans, final_antonyms_ans, final_homophone_ans, final_compound_ans, final_meaning_ans, final_sentence_ans]


  file_name_len = len(excel_file) - 5
  final_file_output = excel_file[0:file_name_len] + " " + col_name + " output" + '.xlsx'

  writer = pd.ExcelWriter(final_file_output)
  for z, df in enumerate (all_df):
      if z == 0:
          name = 'Synonyms'
      elif z==1:
          name = 'Antonyms'
      elif z==2:
          name = 'Homophones'
      elif z==3:
          print(df)
          name = 'Compound words'
      elif z==4:
          name = 'Meanings'
      elif z==5:
          name = 'Sentence'
      df.to_excel(writer, sheet_name = name, index=False)

  writer._save()
  files.download(final_file_output)
  column_index +=1

print("Process finished")
```
